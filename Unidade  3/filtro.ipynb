```python
# Importando as bibliotecas padrão
import matplotlib.pyplot as plt
import numpy as np

# Criando um filto no domínio do tempo:
t0 = 0.01
tf = 310
dt = 0.001
t = np.arange(t0, tf, dt)
filtro = (np.sin(0.2 * np.pi *(t - 155)))/(np.pi *(t - 155))*(0.5 - 0.5* np.cos(2 * np.pi * t / 310))

# Criando um sinal polinomial
#signal = -0.0001*(t-0)*(t-250)

# Convolução entre sinal e o filtro:
#sinal_filtrado = np.convolve(signal, filtro, 'same')

# Criando a figura
plt.figure(1, figsize=(10, 5))
# Realiza a transformada do filtro
plt.subplot(1, 2, 1)
plt.plot(t, filtro, label='filtro')
#plt.plot(t, signal, label='sinal')
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude')
plt.title('filtro de onda senoidal')
plt.legend()
plt.grid()
'''
# Realiza a transformada do filtro
plt.subplot(1, 2, 2)
plt.plot(t, sinal_filtrado, label='sinal filtrado')
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude')
plt.title('filtro de onda senoidal')
plt.legend()
plt.grid()

# # Plotando o espectro de magnitude
#plt.subplot(2, 1, 2)
#plt.magnitude_spectrum(filtro, Fs=1/dt, scale='dB')
'''
plt.tight_layout()
plt.show()
```
