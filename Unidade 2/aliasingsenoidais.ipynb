# **Aliasing de Sinais Senoidais**

A amostragem de sinais senoidais é um fenômeno em que ocorre a distorção do sinal devido a 
uma taxa insuficiente na amostragem de dados.

Um exemplo abaixo: Qual a taxa de amostraem para um sinal senoidal de frequência de 20 Hz?
Segundo Nyquist, a taxa de amostragem mínima para a recuperação deveria ser de 40 Hz.

**CÓDIGO:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Gerando um sinal "contínuo" de onda senoidal - frequência de 20 Hz
t0 = 0                                          # Tempo inicial
tf = 0.5                                        # Tempo final
dt = 0.001                                      # Tempo de amostragem
t = np.arange(t0, tf, dt)                       # Vetor de tempo
s = np.sin(2 * np.pi * 20 * t)                  # Sinal de onda senoidal

# Amostrando o sinal a uma taxa de amostragem de 40 Hz
freq_40 = 40                                          # Frequência de amostragem
T_40 = 1 / freq_40                                    # Tempo de amostragem
t_sample_40 = np.arange(t0, tf, T_40)                 # Vetor de tempo de amostragem
s_sample_40 = np.sin(2 * np.pi * 20 * t_sample_40)    # Sinal de onda senoidal amostrado

# Amostrando o sinal a uma taxa de amostragem de 80 Hz
freq_80 = 80                                          # Frequência de amostragem
T_80 = 1 / freq_80                                    # Tempo de amostragem
t_sample_80 = np.arange(t0, tf, T_80)                 # Vetor de tempo de amostragem
s_sample_80 = np.sin(2 * np.pi * 20 * t_sample_80)    # Sinal de onda senoidal amostrado

# Amostrando o sinal a uma taxa de amostragem de 180 Hz
freq_180 = 180                                          # Frequência de amostragem
T_180 = 1 / freq_180                                    # Tempo de amostragem
t_sample_180 = np.arange(t0, tf, T_180)                 # Vetor de tempo de amostragem
s_sample_180 = np.sin(2 * np.pi * 20 * t_sample_180)    # Sinal de onda senoidal amostrado

plt.figure(1, figsize=(10, 5))
# Plotando o sinal de onda senoidal
plt.subplot(2, 1, 1)
plt.plot(t, s)
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude')
plt.title('Sinal de onda senoidal')
plt.grid()

# Plotando o sinal de onda senoidal amostrado
plt.subplot(2, 1, 2)
plt.plot(t_sample_40, s_sample_40)
plt.plot(t_sample_80, s_sample_80)
plt.plot(t_sample_180, s_sample_180)
plt.ylim(-1.1, 1.1)
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude')
plt.title('Sinal de onda senoidal amostrado')
plt.grid()
plt.legend(['40 Hz', '80 Hz', '180 Hz'])

plt.tight_layout()
plt.show()
```

Sempre existe alguma perda em um sinal amostrado. Contudo, oteorema de Nyquist-Shannon nos diz que o espectro de um sinal não será afetado pela sua discretização se esta ocorrer com uma frequência de amostragem duas vezes maior que a maior frequência do sinal. Amostragem que não respeita o limite de Nyquist pode distorcer o sinal

![](../Imagens/Unidade2/aliasingsenoidais.png)

Para calcular a frequência alias devido a amostragem que não seguiu o  critério de Nyquist é necessário usar a equação abaixo:

$$f_a(N) = f_{in} - N \cdot f_s$$

onde 
* $f_a$ é a frequência de aliasing.
* $f_s$ é a taxa de amostragem.
* $f_{in}$ é a frequência do sinal de entrada.
